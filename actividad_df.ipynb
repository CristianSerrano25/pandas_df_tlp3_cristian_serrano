```python
import pandas as pd
import sqlite3

#Aqui agrega la conexión a la base de datos
conn = sqlite3.connect('Salaries.sqlite')

#Crea el dataframe con los datos del datasets
df = pd.read_sql_query("SELECT * FROM Salaries", conn)
conn.close()

df.head()
```

1. **Top 10 empleados con mayor salario total (incluyendo beneficios)**
Extrae e imprime un top 10 de expleados con mayor salario incluyendo los beneficios 


```python
#Ordena el dataframe por la columna TotalPayBenefits de forma descendente.
df.sort_values(by="TotalPayBenefits", ascending=False, inplace=True) 
#Elimina la columna index que se genera al ordenar el dataframe.
df.reset_index(drop=True, inplace=True)
#Muestra las 10 primeras filas del dataframe con las columnas EmployeeName y TotalPayBenefits.
df[["EmployeeName", "TotalPayBenefits"]].head(10)
```

2. **Filtrar empleados con más de 50,000 en horas extra, la columna a tratar es: (OvertimePay)**

```python
# Convierte la columna OvertimePay a tipo numerico.
df["OvertimePay"] = pd.to_numeric(df["OvertimePay"], errors="coerce")

# Filtra empleados con mas de 50k horas extras e imprime los primeros 10 resultados.
df[df["OvertimePay"] > 50000][["EmployeeName", "OvertimePay"]].head(10) 
```

3. **Contar cuántos empleados únicos hay por año**

```python
# Agrupa el dataframe por la columna Year y cuenta el número de empleados únicos utilizando el metodo .nunique() por año.
df.groupby("Year")["EmployeeName"].nunique()
```

4. **Ver cuántos cargos únicos (JobTitle) existen y los 5 más comunes**

```python
# Imprime el número de cargos únicos en el dataframe.
cargos_unicos = df["JobTitle"].unique()
print(f"La cantidad de cargos únicos es: {len(cargos_unicos)}")

# Se cuentan los valores comunes de "JobTitle" y se imprimen los primeros 5
df["JobTitle"].value_counts().head(5)

```

5. **Mostrar el salario total (TotalPay) promedio por año**

```python
# Se agrupa el dataframe por columna y optiene el promedio de la columna TotalPay por año.
salario_anual_promedio = df.groupby("Year")["TotalPay"].mean()
salario_anual_promedio
```

6. **Exporta todos los datos a un CSV y muestra los ultimos 10 valores**

```python
# Se exporta el dataframe a un archivo CSV llamado "Salaries.csv"

df.to_csv("Salaries.csv", index=False)

df.tail(10)
```
